# Calculate Number of Steps Used

```python
import re
import logging
from tqdm import tqdm
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
import yaml
import numpy as np


logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s [%(levelname)s] %(message)s",
    handlers=[
        logging.FileHandler("process.log"),
        logging.StreamHandler()
    ]
)
logger = logging.getLogger(__name__)


def create_step_listing_prompt(question: str) -> str:
    """
    Formats the prompt to instruct the model to list reasoning steps.
    """
    return (
        "Please provide a detailed, step-by-step solution to the following problem. "
        "Number each step sequentially.\n\n"
        f"Problem:\n{question}\n\n"
        "Solution Steps:"
    )

def generate_step_list_response(
    question: str,
    tokenizer,
    model,
    device,
    max_length: int = 500,
    temperature: float = 0.7,
    top_p: float = 0.95
) -> str:
    """
    Generates a response with a numbered list of steps to solve the problem.
    """
    prompt = create_step_listing_prompt(question)
    input_ids = tokenizer.encode(prompt, return_tensors="pt").to(device)

    try:
        with torch.no_grad():
            output_ids = model.generate(
                input_ids,
                max_length=max_length,
                temperature=temperature,
                top_p=top_p,
                do_sample=True,
                eos_token_id=tokenizer.eos_token_id,
                pad_token_id=tokenizer.eos_token_id
            )
        return tokenizer.decode(output_ids[0], skip_special_tokens=True)
    except Exception as e:
        logger.error(f"Error generating response for question: {question}\nError: {e}")
        return ""


def count_explicit_steps(response: str) -> int:
    """
    Count explicit numbered steps (e.g., 'Step 1', '1.', '1)') in the response.
    """
    if not isinstance(response, str):
        return 0

    pattern = re.compile(r"^\s*(?:Step\s*\d+|\d+[\.\)])", re.MULTILINE | re.IGNORECASE)
    matches = pattern.findall(response)
    return len(set(matches))  # Deduplicate

def count_reasoning_keywords(response: str) -> int:
    """
    Count reasoning keywords as fallback for thought counting.
    """
    if not isinstance(response, str):
        return 0

    keywords = [
        "first", "second", "third", "next", "then",
        "after that", "finally", "therefore", "thus", "so", "hence", "lastly"
    ]
    pattern = re.compile(r"\b(" + "|".join(keywords) + r")\b", re.IGNORECASE)
    return len(pattern.findall(response))

def count_steps_combined(response: str) -> int:
    """
    Combines explicit step counting with keyword-based reasoning detection.
    """
    explicit_steps = count_explicit_steps(response)
    if explicit_steps > 0:
        return explicit_steps
    return count_reasoning_keywords(response)

# ----------------- Save Responses to YAML -----------------
def save_responses_to_yaml(df: pd.DataFrame, yaml_file_path: str = "responses.yaml"):
    """
    Saves the thought process (responses) to a YAML file.
    """
    try:
        responses_dict = {}
        for idx, row in df.iterrows():
            question_number = f"Question_{idx + 1}"
            responses_dict[question_number] = {
                "question": row.get("question", ""),
                "response_small": row.get("response_small", ""),
                "thought_count_small": row.get("thought_count_small", 0),
                "response_medium": row.get("response_medium", ""),
                "thought_count_medium": row.get("thought_count_medium", 0),
                "response_large": row.get("response_large", ""),
                "thought_count_large": row.get("thought_count_large", 0),
                "average_thought_count": row.get("average_thought_count", 0),
            }

        with open(yaml_file_path, "w", encoding="utf-8") as file:
            yaml.dump(responses_dict, file, sort_keys=False, allow_unicode=True)

        logger.info(f"Responses successfully saved to '{yaml_file_path}'.")
    except Exception as e:
        logger.error(f"Failed to save responses to YAML. Error: {e}")
        raise


def display_statistics(df: pd.DataFrame):
    """
    Displays basic statistics on thought counts in the DataFrame.
    """
    average_thoughts = df['average_thought_count'].mean()
    median_thoughts = df['average_thought_count'].median()
    min_thoughts = df['average_thought_count'].min()
    max_thoughts = df['average_thought_count'].max()
    std_thoughts = df['average_thought_count'].std()

    print("\nStatistics on Average Thought Counts:")
    print(f"Average number of thoughts per question: {average_thoughts:.2f}")
    print(f"Median number of thoughts per question: {median_thoughts}")
    print(f"Minimum number of thoughts: {min_thoughts}")
    print(f"Maximum number of thoughts: {max_thoughts}")
    print(f"Standard Deviation: {std_thoughts:.2f}")

# ----------------- Main Processing Function -----------------
def main():
    output_csv = "combined_thoughts.csv"
    output_yaml = "responses.yaml"

    results = []

    # Loop over each question in the DataFrame.
    # Here we assume that the DataFrame has an 'instructions' column.
    for idx, question in tqdm(df['instructions'].items(), total=df.shape[0], desc="Processing Questions"):
        # Generate responses from each of the three models
        response_small = generate_step_list_response(question, tokenizer, model, "cuda")
        response_medium = generate_step_list_response(question, tokenizer, model_medium, "cuda")
        response_large = generate_step_list_response(question, tokenizer, model_large, "cuda")

        # Count the number of thoughts (steps) in each response
        count_small = count_steps_combined(response_small)
        count_medium = count_steps_combined(response_medium)
        count_large = count_steps_combined(response_large)

        # Compute the average thought count across models
        average_thought_count = np.mean([count_small, count_medium, count_large])

        # Append the results for this question
        results.append({
            "question": question,
            "response_small": response_small,
            "thought_count_small": count_small,
            "response_medium": response_medium,
            "thought_count_medium": count_medium,
            "response_large": response_large,
            "thought_count_large": count_large,
            "average_thought_count": average_thought_count
        })

    #  Collect results
    df_results = pd.DataFrame(results)

    # Save the results to a CSV file
    df_results.to_csv(output_csv, index=False)
    print(f"Combined thought counts saved to '{output_csv}'")

    # Save the responses and thought counts to a YAML file
    save_responses_to_yaml(df_results, yaml_file_path=output_yaml)
    print(f"Responses successfully saved to '{output_yaml}'")

    # Display basic statistics on the thought counts
    display_statistics(df_results)

    return df_results



# ----------------- Run the Script -----------------
if __name__ == "__main__":
    df = main()
```
